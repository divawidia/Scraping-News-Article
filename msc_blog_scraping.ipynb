```python
import requests
import pandas as pd
from pymongo import MongoClient
```

**Scraping data produk dari keyword**

```python
blog_list = []
url = "https://www.mscpressarea.com/page-data/en_US/press-releases/page-data.json"

querystring = {}

payload = ""
headers = {
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36 Edg/112.0.1722.39',
        }   
response = requests.request("GET", url, data=payload, headers=headers, params=querystring)

data = response.json()
for blog in data['result']['data']['allDatoCmsMscArticle']['edges']:
        blog_list.append(blog)
```

```python
df = pd.json_normalize(blog_list)
df
```

```python
blog_list2 = []
```

```python
for i in range(0, len(blog_list)):
    blog = blog_list[i]
    url_blog = "https://www.mscpressarea.com/page-data/en_US/press-releases/"+blog_list[i]['node']['slug']+"/page-data.json"

    querystring_review = {}

    payload_review = ""
    headers_review = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36 Edg/112.0.1722.39"
    }

    response_review = requests.request("GET", url_blog, data=payload_review, headers=headers_review, params=querystring_review)

    data_blog = response_review.json()

    for a in data_blog["result"]['data']['datoCmsMscArticle']['body']:
        a.update({"isi_blog" : blog})
        blog_list2.append(a)
    
    print("Scraping blog produk no. "+str(i)+" id: " + blog_list[i]['node']['id'] + " selesai")
```

```python
df = pd.json_normalize(blog_list2)
df
```

```python
df = pd.read_excel('data_blog_msc.xlsx')
df
```

```python
import re
import html

def remove_tags(string): 
    string = re.sub('<.*?>','',string)
    string = html.unescape(string)
    return string

df['text_clean']=df['text'].apply(lambda text : remove_tags(text))
df
```

```python
date1 = pd.Timestamp('2021-01-01T00:00:00.000+00:00')
filtered_df = df.loc[(df['isi_blog.node.meta.createdAt'] >= date1)]
filtered_df
```

```python
filtered_df['isi_blog.node.meta.createdAt'] = pd.to_datetime(filtered_df['isi_blog.node.meta.createdAt'], errors='coerce')
filtered_df['isi_blog.node.meta.createdAt'] = filtered_df['isi_blog.node.meta.createdAt'].dt.tz_localize(None)
```
