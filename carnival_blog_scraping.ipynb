```python
from webdriver_manager.chrome import ChromeDriverManager 
from selenium import webdriver 
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

data_blog = []
browser = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install())) 

for page in range(0, 16):
    browser.get("https://carnival-news.com/news-blog/page/"+str(page)+"/")

    html = browser.page_source

    soup = BeautifulSoup(html, 'html.parser')

    links = soup.find_all('div', 'item__content')
    links

    for link in links:
        date = link.find('span', 'date').text
        title = link.find('a').text
        link_blog = link.find('a').attrs['href']
        data_blog.append([title, date, link_blog])
    print("Data blog page " + str(page) +" sudah dikumpulkan")

for i in range(0, len(data_blog)):
    text = ''
    browser.get(data_blog[i][2])
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')
    artikel = soup.find_all('p')
    for a in artikel:
        text = text + '\n' + str(a)
    data_blog[i].append(text)
    print(i)
```

```python
import pandas as pd
df = pd.DataFrame(data_blog)
df
```

```python
df.rename(columns = {df.columns[0]:'title', df.columns[1]:'date', df.columns[2]:'link', df.columns[3]:'article'}, inplace = True)
df
```

```python
import re
import html

def clean_text(string): 
    string = re.sub('<.*?>','',string)
    string = html.unescape(string)
    return string

df['article']=df['article'].apply(lambda text : clean_text(text))
df
```

```python
df.to_excel('data blog carnival cruise from 2021.xlsx', index=False)
```
