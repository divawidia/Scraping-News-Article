```python
from webdriver_manager.chrome import ChromeDriverManager 
from selenium import webdriver 
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

browser = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install())) 
browser.implicitly_wait(5)
browser.get("https://www.princess.com/news/news_releases/")

html = browser.page_source

soup = BeautifulSoup(html, 'html.parser')

data_blog = []

links = soup.find_all('li', 'article-item')


for link in links:
    date = link.find('div', 'article-date').text
    title = link.find('a').text
    link_blog = link.find('a').attrs['href']
    data_blog.append([title, date, link_blog])
print("Data blog sudah dikumpulkan")

for i in range(0, len(data_blog)):
    text = ''
    browser.get("https://www.princess.com" + data_blog[i][2])
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')
    artikel = soup.find_all('p')
    for a in artikel:
        text = text + '\n' + str(a)
    data_blog[i].append(text)
    print(i)
```

```python
import pandas as pd
df = pd.DataFrame(data_blog)
df
```

```python
import re
import html

def clean_text(string): 
    string = re.sub('<.*?>','',string)
    string = html.unescape(string)
    return string

df['article']=df.iloc[:,3].apply(lambda text : clean_text(text))
df
```

```python
df.drop([df.columns[2], df.columns[3]], axis=1, inplace=True)
df.rename(columns = {df.columns[0]:'title', df.columns[1]:'date'}, inplace = True)
df
```

```python
df.to_excel('data blog princess cruise from 2021.xlsx', index=False)
```
